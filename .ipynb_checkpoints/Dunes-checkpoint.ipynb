```python
def parse_csv(path_to_csv,channel=69):
    """
    path_to_csv: path to file
    channel: spectra of interest, default is 69.
    
    Returns dictionary containing desired data. Only works for * query.
    """
    data = []
    
    with open(path_to_csv,'r') as f:
        
        head = f.readline().replace('"','').strip().split(',')
        while True:
            try:
                entry = f.readline()
            except:
                break
            if entry == '':
                break
                
            entry=entry.replace('"','').split('{')
            info = entry[0][:-1].split(',')
            
            spectra = list(map(float,entry[1][:-2].split(',')))

            entry_dict = dict()
            for i,col in enumerate(head[:-1]):
                if col not in ['name','sequenceid','observationid','starttime']:
                    entry_dict[col]=float(info[i])
                else:
                    entry_dict[col]=info[i]
            entry_dict['channel']=spectra[channel-1]
            entry_dict['spectrum'] = spectra
            data.append(entry_dict)
    return data
        
```

```python
def fit(info):
    """
    info: Dictionary of the format specified in parse_csv
    
    Returns: Gaussian amplitude of fitting routine.
    """
    band_channels = [29,30,31,32,47,48,49,50,51,52,53,54,55,84,85,86,87,88,89,90] + [i for i in range(160,181)] 

    if info['latitude']<-90:
        raise ValueError#, "NOT ON DISK"
    else:
        try:
            p_fit,p_cov = sco.curve_fit(powerlaw,vims_wave[band_channels],info['spectrum'][band_channels],p0=(0.005,-2,0),maxfev=10000)
            p=powerlaw(vims_wave,*p_fit)
            
            #Subtract powerlaw
            adjusted = info['spectrum']-p
            
            #Fit Gaussian
            fit,cov=sco.curve_fit(gaussian,vims_wave[64:85],adjusted[64:85],p0=(2.08,.125,.1),maxfev=100000)
            return fit[-1]
        except:
            #powerlaw fails, directly fit gaussian
            fit,cov = sco.curve_fit(gaussian, vims_wave[64:85],info['spectrum'][64:85],p0=(2.08,.125,.1),maxfev=100000)
            return fit[-1]
```

```python
def adjust(parsed_data):
    """
    parsed_data: Data of the format outlined by parse_csv.
    
    Returns: parsed_data with channel changed to gaussian fit amplitude.
    """
    for i in range(len(parsed_data)):
        parsed_data[i]['channel'] = fit(parsed_data[i])
    
    return parsed_data
    
```

```python
def plot_spectra(data):
    """
    Expects data of the format returned from parse_csv.
    
    Plots spectra vs time.
    """
    
    #When plotting by ettime or Time.jd, the observations are too spaced out, can't see anything. Here just plotted the time strings.
    
    time=[]
    spectra=[]
    phase=[]
    incidence=[]
    emission=[]
    latitude=[]
    longitude=[]
    smallest=None
    largest=None
    for entry in data:
        #time.append(entry['ettime'])
        time.append(entry['starttime'])
        spec = entry['channel']*np.cos(entry['incidence']*np.pi/180)/(np.cos(entry['incidence']*np.pi/180)+np.cos(entry['emission']*np.pi/180))
        #spectra.append(entry['channel'])
        spectra.append(spec)
        smallest=spec if smallest==None or spec<smallest else smallest
        largest=spec if largest==None or largest<spec else largest
        phase.append(entry['phase'])
        incidence.append(entry['incidence'])
        emission.append(entry['emission'])
        latitude.append(entry['latitude'])
        longitude.append(entry['longitude'])
#     fig = plt.figure()
#     ax = fig.add_subplot(111, projection='3d')
#     ax.scatter(phase,incidence,emission)
#     ax.set_xlabel('phase')
#     ax.set_ylabel('incidence')
#     ax.set_zlabel('emission')

    #plt.scatter(phase,spectra)
    #plt.scatter(incidence,spectra)
    #plt.scatter(emission,spectra)
    #plt.scatter(Time(time),spectra)
    
    #UNCOMMENT BELOW TO PLOT
    plt.scatter(time,spectra, c=phase)
    plt.colorbar()
    plt.xticks(rotation="45", fontsize='7')
    plt.xlabel("Time")
    plt.ylabel("Channel")
    plt.title("Spectra vs. Time within Bounding Box")
    #return time,spectra,phase,incidence,emission,latitude,longitude,smallest,largest
```

```python
def create_overlay(query_file,fit=False):
    """
    query_file: name of csv containing query info.
    fit: True if applying fitting routine to spectrum.
    
    Only works for dunes_larger.csv.
    """
    if not fit:
        time,spectra,phase,incidence,emission,latitude,longitude,smallest,largest = plot_spectra(parse_csv(query_file))
    else:
        time,spectra,phase,incidence,emission,latitude,longitude,smallest,largest = (plot_spectra(adjust(parse_csv(query_file))))

    
    dates = {}
    for i,t in enumerate(time):
        d = t[:10]
        if dates.get(d,None) == None:
            dates[d]={'latitude':[latitude[i]],'longitude':[longitude[i]],'spectra':[spectra[i]],'phase':[phase[i]],'incidence':[incidence[i]],'emission':[emission[i]]}
        else:
            dates[d]['latitude'].append(latitude[i])
            dates[d]['longitude'].append(longitude[i])
            dates[d]['spectra'].append(spectra[i])
            dates[d]['phase'].append(phase[i])
            dates[d]['incidence'].append(incidence[i])
            dates[d]['emission'].append(emission[i])
    #return dates
    key = {(0,0): '2007-05-28',(0,1):'2007-05-29',(1,0):'2007-06-13',(1,1):'2007-06-14',(2,0):'2007-06-14',(2,1):'2012-05-22'}
    fig,axes = plt.subplots(3,2)
    for i in range(3):
        for j in range(2):
            axes[i,j].imshow(iss,extent=[0,360,-90,90],cmap='gray')
            axes[i,j].set_xlim(112.5,132.5)
            axes[i,j].set_ylim(-20,5)
            axes[i,j].set_title(key[i,j])
            im=axes[i,j].scatter(dates[key[i,j]]['longitude'],dates[key[i,j]]['latitude'],c=dates[key[i,j]]['spectra'],vmin=smallest,vmax=largest)
            plt.colorbar(im,ax=axes[i,j])

    
    plt.show()
    print(smallest,largest)
    return dates
```
